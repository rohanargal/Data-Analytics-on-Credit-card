```python
import pandas as pd 
```

```python
df1=pd.read_csv("Credit_card.csv")
```

```python
df2=pd.read_csv("Credit_card_label.csv")
```

```python
df=pd.merge(df1,df2,on='Ind_ID')
```

```python
df.isnull().sum()
```

```python
df['GENDER']=df['GENDER'].fillna(df['GENDER'].mode()[0])
```

```python
df['Type_Occupation']=df['Type_Occupation'].fillna(df['Type_Occupation'].mode()[0])
```

```python
df['Annual_income']=df['Annual_income'].fillna(df['Annual_income'].mean())
```

```python
reference_date = pd.to_datetime('2024-01-26')

# Calculate the actual birthdates by adding the relative counts to the reference date
df['birthdate'] = reference_date + pd.to_timedelta(df['Birthday_count'], unit='D')
```

```python
df['birthdate']
```

```python
df['birthdate'] = pd.to_datetime(df['birthdate'])
```

```python
df['days_until_now'] = (pd.to_datetime('now') - df['birthdate']).dt.days
```

```python
df.drop(columns=['Birthday_count'], inplace=True)
```

```python
df['employee_days'] = df['Employed_days'].abs() * df['days_until_now']
```

```python
df.drop(columns=['Employed_days','birthdate'], inplace=True)
```

```python
df['days_until_now']=df['days_until_now'].fillna(round(df['days_until_now'].mean(),0))
```

```python
df['employee_days']=df['employee_days'].fillna(round(df['employee_days'].mean()))
```

```python
import duckdb
con=duckdb.connect()
con.register("df",df)
con.execute("select * from df").fetchdf()
```

1.Group the customers based on their income type and find the average of their annual income.

```python
con.execute("SELECT Type_Income, AVG(Annual_income) AS Avg_Income FROM df GROUP BY Type_Income").fetchdf()
```

2. Find the female owners of cars and property.

```python
con.execute("SELECT * FROM df WHERE GENDER = 'F' AND (Car_Owner = 'Y' AND Propert_Owner = 'Y')").fetchdf()
```

3.Find the male customers who are staying with their families.

```python
con.execute("SELECT * FROM df WHERE GENDER = 'M' AND Marital_status = 'Married' AND Family_Members > 2").fetchdf()
```

4.Please list the top five people having the highest income.

```python
con.execute("SELECT * FROM df ORDER BY Annual_income DESC LIMIT 5").fetchdf()
```

5.How many married people are having bad credit?

```python
con.execute("SELECT COUNT(*) AS Married_Bad_Credit_Count FROM df WHERE Marital_status = 'Married' AND label = 1;").fetchdf()
```

6.What is the highest education level and what is the total count?

```python
con.execute("SELECT EDUCATION, COUNT(*) AS Total_Count FROM df GROUP BY EDUCATION ORDER BY Total_Count DESC LIMIT 1").fetchdf()
```

7.Between married males and females, who is having more bad credit? 

```python
con.execute("SELECT Marital_status, GENDER, COUNT(*) AS Bad_Credit_Count FROM df WHERE label = 1 AND Marital_status = 'Married' GROUP BY Marital_status, GENDER").fetchdf()
```
