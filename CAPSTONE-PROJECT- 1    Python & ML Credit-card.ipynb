Data Preparation has problems to solve like the outliers, encoding techniques, feature selection are done respected methods.

to Check outlier we have to use graphs for easy considerations.

```python
sns.boxplot(df['days_until_now'])
plt.show()
```

```python
sns.boxplot(df['employee_days'])
plt.show()
```

```python
sns.distplot(df['employee_days'])
plt.show()
```

From the above Graphs we see that emplyee_days has more outlier and is not symmetry. 
We  Deal outliers with IQR Method.

```python
Q1=df.describe()['employee_days']['25%']
Q3=df.describe()['employee_days']['75%']
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
df['employee_days']=df['employee_days'].clip(upper_limit,lower_limit)
```

```python
sns.boxplot(df['employee_days'])
plt.show()
```
